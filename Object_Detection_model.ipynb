```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load your model
model = YOLO('yolov8n.pt')

# Run inference
result = model('/content/drive/MyDrive/image detection.jpg')

# Extract the Results object from the list
result = result[0] # Get the first element (Results object)

# Get the image data (plot array) from the Results object
image = result.plot()

# Display the image using Matplotlib
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
```
